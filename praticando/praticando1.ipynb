```python
import pandas as pd

idades = pd.Series([20, 30, 22])

print(idades)
print()
print(idades[2]) # acessando elementos
print()
print(idades + 10) # ele opera em mais de uma coluna por vez

```

```python
tabela = pd.DataFrame({
    "nome": ["Ana", "Carlos", "Maria"],
    "idade": [20, 25, 30]
})

print(tabela)
print()
print(tabela["nome"]) # seleciona apenas a coluna nome
print()
print(tabela["nome"][2])
print()
print(tabela[["nome", "idade"]]) # pegando multiplas colunas
```

```python
# exercicio 2

tabela = pd.DataFrame({
    "nome": ["Ana", "Carlos", "Maria"],
    "nota": [8.5, 7.0, 9.2],
    "faltas": [2, 5, 1]
})

print(tabela["nota"])
print()
linhas, colunas = tabela.shape
print(f"A tabela tem {linhas} linhas e {colunas} colunas.")
print()
print(tabela.head(2))
```

```python
tabela = pd.DataFrame({
    "nome": ["Ana", "Carlos", "Maria", "Pedro"],
    "idade": [20, 25, 30, 18],
    "salario": [2500, 4000, 5000, 1800]
})

# iloc trabalha com posições

print(tabela)
print()
print(tabela.iloc[0]) # pega toda a primeira linha
print()
print(tabela.iloc[:2]) 
```

Enquato iloc trabalha com a posição o loc trabalha com o rótulo do indice

```python
tabela = pd.DataFrame({
        "nome": ["Ana", "Carlos", "Maria"],
        "nota": [8.5, 7.0, 9.2]
    },
    index = ["a", "b", "c"])

print(tabela.loc["b"]) # pega a linha pelo rotulo
print() 
print(tabela.iloc[1]) # pega pela posição
```
