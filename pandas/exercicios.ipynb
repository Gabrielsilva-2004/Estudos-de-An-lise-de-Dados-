## Exercícios 1
Básico

Crie uma Series com os valores:

10, 20, 30, 40

e:

Mostre a Series.<br>
Mostre apenas o primeiro elemento.<br>
Some 5 a todos os elementos.

```python
import pandas as pd

dados1 = pd.Series([10, 20, 30, 40])

print(dados1)
print()
print(dados1.iloc[0])
print()
dados1 = dados1 + 5
print(dados1)
```

## Exercícios 2

Crie a seguinte Series:

notas = pd.Series(
    [8.5, 7.0, 9.2],
    index=["Ana", "Carlos", "Maria"]
)

Faça:

Mostrar a nota de Maria.<br>
Aumentar todas as notas em 0.5.<br>
Verificar quais notas são maiores que 8.

```python
import pandas as pd
notas = pd.Series(
    [8.5, 7.0, 9.2], 
    index=["Anna", "Carlos", "Maria"]
    )

print(notas)
print()
print("Notas de Maria", notas.loc["Maria"])
notas += 0.5
print()
print(notas)
print()
notas_maiores_que_8 = notas > 8
print("Notas maiores que 8")
print(notas_maiores_que_8)
```

## Exercicio 3

Desafio

Sem usar for, crie:

vendas = pd.Series(
    [100, 250, 400, 150],
    index=["Jan", "Fev", "Mar", "Abr"]
)

Depois:

Aumente todas as vendas em 10%.<br>
Mostre apenas os meses com vendas acima de 200.<br>
Calcule a soma total das vendas usando um método do Pandas.

Tente resolver e eu corrijo como um professor, explicando cada ponto.

```python
vendas = pd.Series(
    [100, 250, 400, 150],
    index=["Jan", "Fev", "Mar", "Abr"]
)

aumento_de_10_porcento = vendas * 1.10
print(aumento_de_10_porcento)
print()
vendas_acima_de_200 = vendas[vendas > 200]
print(vendas_acima_de_200)
print()
print("Soma total dos valores:", vendas.sum())
```
